```python
import jsonlines
```

```python
def get_data(split='train'):
    with jsonlines.open('snli_1.0_{}.jsonl'.format(split)) as reader:
        for obj in reader:
            yield obj
```

```python
a=iter(get_data())
```

```python
neutral = open('dev/neutral.txt','w')
entail = open('dev/entailment.txt','w')
contra = open('dev/contradiction.txt','w')
# neutral=[]
# entail=[]
# contra=[]
for obj in get_data('dev'):
#     obj = next(a)
    sentences=[obj['sentence1'],obj['sentence2']]
    if obj['gold_label']=='neutral':
        neutral.write(sentences[0]+'\n'+sentences[1]+'\n')
    elif obj['gold_label']=='entailment':
        entail.write(sentences[0]+'\n'+sentences[1]+'\n')
    elif obj['gold_label']=='contradiction':
        contra.write(sentences[0]+'\n'+sentences[1]+'\n')
    else:
        print(obj['gold_label'])
#     break

neutral.close()
entail.close()
contra.close()
```
